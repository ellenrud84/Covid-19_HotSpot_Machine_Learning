# Forming model_daily_data Table

```python
# Install Dependencies
import pandas as pd
```

## Import .csv Files

```python
# datasets filepaths
lstm_daily_file="lstm_daily_data.csv"
lstm_rmse_file="lstm_rmse.csv"
sarimax_daily_file="sarimax_daily_data.csv"
sarimax_rmse_file="sarimax_rmse.csv"
```

```python
# read datasets
lstm_daily=pd.read_csv(lstm_daily_file, parse_dates=['date'])
lstm_rmse=pd.read_csv(lstm_rmse_file)
sarimax_daily=pd.read_csv(sarimax_daily_file, parse_dates=['date'])
sarimax_rmse=pd.read_csv(sarimax_rmse_file)
```

## Combine Data

### LSTM and SARIMAX Case , Death , and Residual Data

```python
lstm_daily.info()
```

```python
lstm_daily.set_index(['date','fips_code'], inplace = True)
```

```python
lstm_daily.head()
```

```python
sarimax_daily.info()
```

```python
sarimax_daily.set_index(['date','fips_code'], inplace = True)
```

```python
sarimax_daily.head()
```

```python
model_daily_data = lstm_daily.join(sarimax_daily, how='outer')
model_daily_data
```

```python
lstm_rmse.info()
```

```python
lstm_rmse.set_index('fips_code', inplace = True)
```

```python
sarimax_rmse.info()
```

```python
sarimax_rmse.set_index('fips_code', inplace = True)
```

```python
model_rmse_data = lstm_rmse.join(sarimax_rmse, how='outer')
model_rmse_data
```

## Export to CSV

```python
model_daily_data.to_csv('model_daily_data.csv', index=True)
```

```python
model_rmse_data.to_csv('model_rmse_data.csv', index=True)
```
